```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from erftools.input_sounding import InputSounding
```

# compare python sounding with George Bryan's original `mbm.F` code output

```python
orig = InputSounding('mbm.input_sounding')
```

```python
new = InputSounding('BryanFritsch2002_moist_sounding')
```

```python
fig,axs = plt.subplots(ncols=2,sharey=True)

axs[0].plot(orig.th, orig.z, label='mbm.F')
axs[0].plot(new.th, new.z, label='erf')
axs[0].set_xlabel(r'$\theta$ [K]')

axs[1].plot(orig.qv, orig.z, label='mbm.F')
axs[1].plot(new.qv, new.z, label='erf')
axs[1].set_xlabel(r'$q_v$ [kg/kg]')

axs[0].set_ylabel(r'$z$ [m]')
for ax in axs:
    ax.grid()
```
